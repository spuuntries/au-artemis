# Display pickled output

```python
max_items = 10


from IPython.display import display
import pandas as pd
    
from PIL import Image

df= pd.read_csv(csv_out_file)

print(df.columns)

#limit the output:
df = df[:max_items]

count=0

for index in df.index:
    print(index)
    image_file = df['image_file'][index]
    #emotion = df['grounding_emotion'][index]
    caption = df['caption'][index]

    
    ranked_emotions =  [str( df['emo' + str(i)][0])  for i in range(len(IDX_TO_EMOTION))]  
    emotion = ranked_emotions[0]

    img = Image.open(image_file)

    display(img)
    print("main emotion: ", emotion)
    print("caption: ", caption)
    print("ranked emotions:", ",".join(ranked_emotions))
    count+=1
    if count > max_items: break
```
